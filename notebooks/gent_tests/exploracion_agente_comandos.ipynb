```python
# 🔧 Configuración del entorno
import os
import sys
from pathlib import Path
import pandas as pd

# Subir desde notebooks/gent_tests/ u otra carpeta anidada
os.chdir("../../")  # Ajusta este nivel si estás en otra profundidad

# Establecer el directorio raíz del proyecto
project_root = Path().resolve()
print(f"📁 Directorio raíz del proyecto: {project_root}")

# Agregar al sys.path si no está
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
    print("✅ Ruta añadida al sys.path")

# 🔍 Importar el agente
from agent_modules.command_agent import CommandAgent

# 🧠 Entradas de prueba
inputs = [
    "Haz un resumen del archivo",
    "Busca información en la base de datos",
    "Genera un reporte de ventas",
    "Convierte el texto en una lista de tareas",
    "Crea un resumen ejecutivo"
]

# 🚀 Crear agente y ejecutar pruebas
agent = CommandAgent()
resultados = []

for entrada in inputs:
    salida = agent.act(entrada)
    resultados.append({"🧠 Entrada": entrada, "✅ Salida del agente": salida})

# 🧾 Mostrar resultados en tabla
df_resultados = pd.DataFrame(resultados)
df_resultados
```

```python
# Lista de entradas problemáticas
inputs_edge = [
    "",  # Vacío
    "    ",  # Solo espacios
    None,  # Nulo
    123,  # Número en vez de texto
    "🔥🚀💥",  # Emojis
    "SELECT * FROM users;",  # Inyección SQL
    "¿Puedes hacerme café?",  # No relacionado
]

# Recolectar resultados
resultados_edge = []
for entrada in inputs_edge:
    try:
        salida = agent.act(entrada)
        resultados_edge.append({
            "🧪 Entrada atípica": str(entrada),
            "✅ Salida del agente": salida
        })
    except Exception as e:
        resultados_edge.append({
            "🧪 Entrada atípica": str(entrada),
            "❌ Error": str(e)
        })

# Mostrar como DataFrame
df_edge = pd.DataFrame(resultados_edge)
df_edge
```
