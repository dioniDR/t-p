```python
# Celda revisada: Verificar la importaci√≥n de m√≥dulos
import os
import sys
from pathlib import Path

# Obtener el directorio ra√≠z del proyecto
os.chdir("../../")
project_root = Path().resolve()
print(f"Directorio ra√≠z del proyecto: {project_root}")

# A√±adir directorio ra√≠z al path de Python
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
    print(f"A√±adido al path: {project_root}")

# Verificar si existe el directorio agent_modules
agent_modules_dir = project_root / "agent_modules"
if not agent_modules_dir.exists():
    print(f"\n‚ùå El directorio agent_modules no existe en {project_root}")
else:
    print(f"\n‚úÖ El directorio agent_modules existe en {project_root}")
    
    # Listar archivos en agent_modules
    print("\nArchivos en agent_modules:")
    for item in agent_modules_dir.iterdir():
        if item.is_dir():
            print(f"üìÅ {item.name}")
        else:
            print(f"üìÑ {item.name}")

# Verificar si existe el archivo __init__.py con el nombre correcto
init_file = agent_modules_dir / "__init__.py"
wrong_init_file = agent_modules_dir / "**init**.py"

if init_file.exists():
    print(f"\n‚úÖ El archivo __init__.py existe correctamente")
elif wrong_init_file.exists():
    print(f"\n‚ö†Ô∏è Encontrado archivo '**init**.py' en lugar de '__init__.py'")
    print(f"Para corregirlo, ejecuta este comando en la terminal:")
    print(f"mv '{wrong_init_file}' '{init_file}'")
else:
    print(f"\n‚ùå No se encontr√≥ el archivo __init__.py")

# Intentar importar los m√≥dulos
try:
    print("\nIntentando importar m√≥dulos...")
    
    # Alternativa 1: Con el path ajustado
    import agent_modules.base_agent
    import agent_modules.agent_registry
    
    print("\n‚úÖ M√≥dulos importados correctamente")
    
    # Importar las clases espec√≠ficas
    from agent_modules.base_agent import BaseAgent
    from agent_modules.agent_registry import AgentRegistry
    
    print("‚úÖ Clases importadas correctamente")
    print(f"BaseAgent es una clase abstracta: {hasattr(BaseAgent, '__abstractmethods__')}")
    
    # Verificar los m√©todos abstractos requeridos
    if hasattr(BaseAgent, '__abstractmethods__'):
        print(f"M√©todos abstractos: {BaseAgent.__abstractmethods__}")
    
    # Verificar el registro de agentes
    print(f"\nAgentRegistry tiene diccionario de agentes: {hasattr(AgentRegistry, '_agents')}")
    print(f"Agentes registrados: {len(AgentRegistry.get_all_agents())}")
    
except ImportError as e:
    print(f"\n‚ùå Error al importar m√≥dulos: {e}")
    print("Verifica que los archivos est√©n correctamente nombrados:")
    print("- agent_modules/__init__.py (con doble guion bajo)")
    print("- agent_modules/base_agent.py")
    print("- agent_modules/agent_registry.py")
```
